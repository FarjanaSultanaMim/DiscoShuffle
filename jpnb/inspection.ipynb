# Essay Scoring

## DI-aware

```python
def inspect(m, t = "No description", ax = None, vals = "loss val_loss".split()):
    l = pickle.load(open("../{}/logs.pickle".format(m), "rb"))
    p = dict([p.strip().split("=", 1) for p in open("../{}/param.txt".format(m))])
    df_b, df_e = pd.DataFrame(l["logs_b"]), pd.DataFrame(l["logs_e"])
    df_e[vals].plot(ylim=[0, 1], title=t, ax=ax)
    print(p)
    print(df_e[vals].describe())
```

```python
# This setting is weird because DI embeddings are fixed
fig, axs = plt.subplots(1, 3, figsize=(12, 4))
inspect("output/46392cffc6de88b365408aeb30f5415b", "DI-aware, no pretraining (dropout=0.7)", ax=axs[0])
inspect("output/ebb929454385723f06b415f9164590c6", "DI-aware, no pretraining (dropout=0.5)", ax=axs[1])
inspect("output/6640830a05efb4a9dc8eebd1a1e79706", "DI-aware, no pretraining (dropout=0.3)", ax=axs[2])
```
